```python
plt.style.use('seaborn-white')
data=np.random.randn(1000)
plt.hist(data)
```

```python
x1=np.random.normal(0,0.7,1000)#empieza en cero, std =0.7
x2=np.random.normal(-2,1,1000)
x3=np.random.normal(3,1.6,1000)
d=dict(histtype='stepfilled',alpha=0.3,bins=40)
plt.hist(x1,**d)
plt.hist(x2,**d)
plt.hist(x3,**d)
plt.show()
```

```python
#Histogramas bidimencionales
mean = [0,0]
cov =[[1,1],[1,2]]
x,y=np.random.multivariate_normal(mean,cov,1000).T
plt.hist2d(x,y,bins=30,cmap='Blues')
cb=plt.colorbar()
cb.set_label('total por contenedor')
```

```python
plt.hexbin(x,y,bins=30,cmap='Blues') #hexagonitos pegados unos con otros
cb=plt.colorbar()
cb.set_label('total por contenedor')
```
