# Ejemplos matplotlib.pyplot

Existen muchos paquetes de visualización en python, pero el principal es matplotlib. Necesitaremos su subpaquete [pyplot](https://matplotlib.org/stable/api/pyplot_summary.html).
Por convención, este subpaquete se importa como plt:

```python
import matplotlib.pyplot as plt

#También usaremos otras librerías
import numpy as np
```

En el primer ejemplo, podemos tomar algunos datos ficticios sobre ventas de una concesionaria.

```python
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
sales = [100,120,125,128,136, 45, 68, 80, 110, 73]
```

## Plot
Para representar estos datos en forma de gráfico de líneas (o puntos), llamamos a `plt.plot` y utilizamos nuestras dos listas como argumentos. El primer argumento corresponde al eje horizontal y el segundo, al vertical.

```python
# plot traza un gráfico de líneas. Le dice a python qué y cómo
plt.plot(year, sales)
# show muestra el gráfico
plt.show()
```

Podemos agregar características a la gráfica: títulos, cuadrícula, color de línea

```python
# Modificamos el tamaño de la figura
plt.figure(figsize=(8, 5))
# Añadimos una cuadrícula
plt.grid()

plt.plot(year, sales, color='red')
#Añadimos un título
plt.title('Autos vendidos por año.')
# Añadimos título a los ejes
plt.xlabel('Año')
plt.ylabel('Cantidad de autos vendidos')
# Cambiamos los límites de los ejes
plt.xlim((2014, 2025))
#Mostramos los valores deseados en la cuadrícula
plt.xticks(np.arange(2014,2026,1))
# show muestra el gráfico
plt.show()
```

Podemos ver otro ejemplo de la documentación de matplotlib para conocer algunos parámetros que pueden modificarse.

```python
plt.plot([0, 1, 2], [1.5, 2, 3], 'go-', label='line 1', linewidth=2)
plt.plot( [1, 4, 9], 'rs', label='line 2')
plt.show()
```

Podemos mostrar más de una línea por gráfico y modificar más su aspecto.

```python
# Consideramos más datos ficticios

year = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]
sales_honda = [100,120,125,128,136, 45, 68, 80, 110, 62]
sales_suzuki= [112,115,127,132,133, 20, 40, 110, 138, 74]
sales_toyota= [107, 122, 126, 135, 134, 18, 57, 91, 145, 76]
```

```python
# Añadimos una cuadrícula
plt.grid()
# Agregamos características para cada línea
plt.plot(year, sales_honda,
         color='g', marker='o', linestyle='dashed',
         linewidth=1.6, markersize=7, label = 'Honda')
plt.plot(year, sales_suzuki,
         c=(0.1, 0.2, 0.5), marker = '*', linestyle = '--',
         label = 'Suzuki')
plt.plot(year, sales_toyota,
         color='orangered', marker = 'v', linestyle = ':',
         label = 'Toyota')

plt.title('Autos vendidos por año, por marca.')
plt.xlabel('Año')
plt.ylabel('Cantidad de autos vendidos')
#Mostramos los valores deseados en la cuadrícula
plt.xticks(np.arange(2015,2025,1))
# Hacemos que se muestre el cuadro de nombres
plt.legend()
plt.show()
```

## Histograma

Para estos mismo datos, podemos obtener un histograma con `hist`.

```python
# histograma
plt.hist(sales_honda)
plt.show()
```

```python
# Nuevamente, podemos modificar los parámetros para personalizar la gráfica

plt.grid(alpha=0.4)
plt.hist([sales_honda, sales_suzuki, sales_toyota], bins=3,
         label = ['Honda', 'Suzuki', 'Toyota'],
         color=['deeppink', 'orangered', 'crimson'])
plt.legend(prop={'size': 10})
plt.show()
```

Veamos un ejemplo con una distribución conocida.

```python
# tomamos valores aleatorios a partir de la distribución normal estándar
xdata = np.random.normal(size=1000)
# definimos el soporte de la densidad real
xpdf = np.arange(-4, 4, 0.1)
# y la función de densidad de probabilidad
pdf = 1 / (np.sqrt(2 * np.pi)) * np.exp(-xpdf**2 / 2)
```

```python
# Definimos los intervalos para el histograma
dx = 0.1 #0.4
xbins = np.arange(-4, 4, dx)
```

```python
plt.hist(xdata, bins=xbins, density=False, histtype='step', label='Counts')
# Para escalar y graficar la pdf:
# pdf * len(xdata) * dx convierte la pdf continua en una forma que representa el número esperado de recuentos por casilla,
#   que puede compararse directamente con los recuentos de su histograma.
plt.plot(xpdf, pdf * len(xdata) * dx, label=r'Distribución no-normalizada')
plt.ylabel('Count per bin')
plt.xlabel('x bins')
plt.legend(loc='lower left')
plt.show()
```

## Scatter

Otro tipo de gráfico es el de dispersión, que podemos visualizar con `plt.scatter`. Este gráfico muestra la posición de puntos dentro del plano bidimensional. Para el ejemplo tomaremos datos sintéticos.

```python
# Se establece una semilla para tener reproducibilidad
np.random.seed(3)


N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors_1 = np.random.rand(1)
colors_2 = np.random.rand(1)
area = (30 * np.random.rand(N))**2  # 0 to 15 point radii

# s: tamaño del punto
# alpha: transparencia
plt.scatter(x, y, s=area, c=[colors_1]*25+[colors_2]*25, alpha=0.5)

plt.title('Dispersión de puntos')
plt.show()
```

# Subplots

En el siguiente ejemplo además usaremos [`plt.subplots()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html)

"Crea una figura y un conjunto de *subplots*.

Este *utility wrapper* hace que sea fácil crear diseños comunes de *subplots*, incluyendo el objeto figura que encierra, en una sola llamada."

```python
# Número de puntos
N = 45
#Coordenadas x , y para los N puntos
x, y = np.random.rand(2, N)

# Escogemos 5 colores para definir clases, dados por números entre 1 y 5
c = np.random.randint(1, 5, size=N)
# También el tamaño será aleatorio
s = np.random.randint(10, 220, size=N)

# Para hacer subplots, suele usarse el método como sigue
fig, ax = plt.subplots()

# definimos un objeto scatter
scatter = ax.scatter(x, y, c=c, s=s)

# Hace una leyenda con los colores de la dispersión; indica la posición
legend1 = ax.legend(*scatter.legend_elements(),
                    loc="lower left", title="Classes")
# add_artist: Clase base abstracta para objetos que se renderizan en un FigureCanvas.
ax.add_artist(legend1)

# produce a legend with a cross-section of sizes from the scatter
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.4)
legend2 = ax.legend(handles, labels, loc="upper right", title="Sizes")

plt.show()
```

Aquí hay más ejemplos de `subplots`: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#examples-using-matplotlib-pyplot-subplots

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

x = np.random.rand(10)
y = np.random.rand(10)
z = np.sqrt(x**2 + y**2)

fig, axs = plt.subplots(2, 3, sharex=True, sharey=True, layout="constrained")

# Matplotlib marker symbol
axs[0, 0].scatter(x, y, s=80, c=z, marker=">")
axs[0, 0].set_title("marker='>'")

# marker from TeX: passing a TeX symbol name enclosed in $-signs
axs[0, 1].scatter(x, y, s=80, c=z, marker=r"$\clubsuit$")
axs[0, 1].set_title(r"marker=r'\$\clubsuit\$'")

# marker from path: passing a custom path of N vertices as a (N, 2) array-like
verts = [[-1, -1], [1, -1], [1, 1], [-1, -1]]
axs[0, 2].scatter(x, y, s=80, c=z, marker=verts)
axs[0, 2].set_title("marker=verts")

# regular pentagon marker
axs[1, 0].scatter(x, y, s=80, c=z, marker=(5, 0))
axs[1, 0].set_title("marker=(5, 0)")

# regular 5-pointed star marker
axs[1, 1].scatter(x, y, s=80, c=z, marker=(5, 1))
axs[1, 1].set_title("marker=(5, 1)")

# regular 5-pointed asterisk marker
axs[1, 2].scatter(x, y, s=80, c=z, marker=(5, 2))
axs[1, 2].set_title("marker=(5, 2)")

plt.show()
```

Vemos otro ejemplo con `plt.bar`.

```python
species = ('Adelie', 'Chinstrap', 'Gentoo')
sex_counts = {
    'Male': np.array([73, 34, 61]),
    'Female': np.array([73, 34, 58]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(3)

for sex, sex_count in sex_counts.items():
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    ax.bar_label(p, label_type='center')

ax.set_title('Number of penguins by sex')
ax.legend()

plt.show()
```
